# 9 . average weekly response time for each agent 

```python
d2.head(2)
```

```python
#d2.groupby(['Agent Name','data_week'])['Average Response Time'].mean()
```

```python
def sec_calculate(time_str):
    
    '''This function will give you your total sececonds of your time.'''
    
    # convert date in hh:mm:ss
    hh,mm,ss = time_str.split(':')
    time = int(hh) * 3600 + int(mm) * 60 + int(ss)
    return time
    
    
    
```

```python
l = []
for i in d2['Average Response Time']:
    l.append(sec_calculate(i))
d2['Average Response Seconds'] = l
```

```python
d2
```

```python
d2.groupby(['Agent Name', 'data_week'])['Average Response Seconds'].mean().to_frame()
```
